```python
#Importing seaborn
import seaborn as sns
```

```python
#Importing other libraries also
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Total bill v/s tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#setting theme 
sns.set_theme()
```

```python
#Visualize the data
sns.relplot(data=tips, x='total_bill', y='tip',col='time', hue='smoker',style='smoker',size='size')
```

```python
#Loading the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Iris dataset
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
#Loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(x='class',data=titanic)
```

```python
titanic.shape
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
#House price dataset
boston_dataset = pd.read_csv('BostonHousing.csv')
boston_df = pd.DataFrame(boston_dataset)
boston_df.head()
```

```python
sns.distplot(boston_df['age'])
```

```python
#constructing a heatmap
correlation = boston_df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
