```python
 #Importing matplotlib 
import matplotlib.pyplot as plt
```

```python
#importing numpy as well as pandas 
import numpy as np
import pandas as pd
```

```python
#generating random values for our plot
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
x
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title, x-axis & y-axis to the plot
plt.plot(x,y)
plt.xlabel('ANGLE')
plt.ylabel('SINE VALUE')
plt.title('SINE PLOT GRAPH')
plt.show()
```

```python
#Parabloa
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
#plotting with different color and symbol
x = np.linspace(-40,40,50)
y = x**2
plt.plot(x,y,'r+')  #r-> red  +->symbol
plt.title('COLOR AND SYMBOL CHANGE')
plt.show()
```

```python
x = np.linspace(-40,40,50)
y = x**2
plt.plot(x,y,'gx')  #g -> green  x ->symbol
plt.title('COLOR AND SYMBOL CHANGE')
plt.show()
```

```python
x = np.linspace(-5,5,20)
y = x**2
plt.plot(x,y,'y.')  #y -> yello  . ->symbol
plt.title('COLOR AND SYMBOL CHANGE')
plt.show()
```

```python
#Plotting more than 1 plots in single graph
x = np.linspace(-9,9,40)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

```python
figure = plt.figure()
ax = figure.add_axes([0,0,1,1])
lang = ['English','Hindi','Spanish','Itlaian','German']
people = [120,50,70,20,100]
ax.bar(lang, people)
plt.title('BAR PLOT')
plt.xlabel('LANGUAGE')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
figure = plt.figure()
ax = figure.add_axes([0,0,1,1])
lang = ['English','Hindi','Spanish','Itlaian','German']
people = [120,50,70,20,100]
ax.pie(people, labels = lang, autopct = '%1.1f%%')
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z, color='b')
plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
x = 20 * np.random.random(100)
y = np.sin(x)
z = np.cos(y)
ax.scatter(x,y,z,c=x,cmap='Blues')
plt.show()
```
