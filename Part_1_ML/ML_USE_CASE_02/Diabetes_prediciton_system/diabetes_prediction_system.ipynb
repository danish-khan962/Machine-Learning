#DIABETES PREDICTION SYSTEM

```python
#Importing dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Major and main dependencies
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm 
from sklearn.metrics import accuracy_score
```

##DATA COLLECTION & ANALYSIS

```python
#Loading the dataset into pandas DataFrame
diabetes_dataset = pd.read_csv('diabetes.csv')
diabetes_df = pd.DataFrame(diabetes_dataset)
```

```python
#printing 5 rows
diabetes_df.head()
```

```python
#checking shape of DataFrame
diabetes_df.shape
```

```python
#checking for missing values
diabetes_df.isnull().sum()
```

```python
#0-> non diabetic
#1-> diabetic
```

```python
#checking for numbers 
diabetes_df['Outcome'].value_counts()
```

```python
#getting the statsitical measures of data
diabetes_df.describe()
```

```python
#getting mean of label 0 and 1 in outcomes column
diabetes_df.groupby('Outcome').mean()
```

```python
#separating data and labels
X = diabetes_df.drop(columns='Outcome',axis=True)
Y = diabetes_df['Outcome']
```

```python
X
```

```python
Y
```

##DATA STANDARDIZATION: Standardizing the data 

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print("Standardized data-> ", standardized_data)
```

```python
#Feeding standardized data to X
X = standardized_data
Y = diabetes_df['Outcome']
```

```python
X
```

```python
Y
```

##DATA SPLITTING : train_test_split

```python
#training and testing our data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
#checking shapes of all transformed data
print(X.shape, X_train.shape, X_test.shape)
```

##TRAINING THE MODEL

```python
#training the model
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector machine(svm) classifier
classifier.fit(X_train, Y_train)
```

##EVALUATING MODEL

```python
#Evaluating model
#Accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Traning data accuracy: ", training_data_accuracy)
```

```python
#Accuracy score on the testing data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Test data accuracy: ", test_data_accuracy)
```

#MAKING A PREDICTIVE SYSTEM

```python
#making predictive system
```

```python
#getting input data
input_data = (2,197,70,45,543,30.5,0.158,53)

#converting input_data into numpy array
input_data_as_array = np.asarray(input_data)

#reshaping numpy array for one instance
input_data_reshaped = input_data_as_array.reshape(1,-1)

#standardizing the input_data
std_data = scaler.transform(input_data_reshaped)

prediction = classifier.predict(std_data)
print("Prediction: ", prediction)
if(prediction[0]==0):
    print("Patient is : NON-DIABETIC")
if(prediction[0]==1):
    print("Patient is: DIABETIC")
```
