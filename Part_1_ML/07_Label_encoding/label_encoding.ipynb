```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
```

```python
#importing important dependncy
from sklearn.preprocessing import LabelEncoder
```

```python
cancer_dataset = pd.read_csv('cancer.csv')
cancer_df = pd.DataFrame(cancer_dataset)
```

```python
cancer_df.head()
```

```python
#Finding the count of different labels
cancer_df['diagnosis'].value_counts()
```

```python
#load the LabelEncoder function
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_df.diagnosis)
```

```python
#appending the labels to DataFrame
cancer_df['target'] = labels
```

```python
cancer_df.head()
```
